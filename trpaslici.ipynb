```python
radek = 2          # 0-index (první řádek)
sloupec = "7"      # název sloupce

# hodnota buňky
hodnota = df.loc[radek, sloupec]
print("Hodnota:", hodnota)

# funkce pro vybarvení jedné buňky
def zvyrazni_bunku(dataframe, r, s):
    def style_func(x):
        df_style = pd.DataFrame("", index=dataframe.index, columns=dataframe.columns)
        df_style.loc[r, s] = "background-color: #4CAF50; color: white"
        return df_style
    return dataframe.style.apply(style_func, axis=None)

# zobrazení tabulky se zelenou buňkou
zvyrazni_bunku(df, radek, sloupec)

def vybarvi_3x3(data, radek, sloupec):
    styles = pd.DataFrame("", index=data.index, columns=data.columns)

    r = radek
    s = int(sloupec)

    for dr in [-1, 0, 1]:
        for ds in [-1, 0, 1]:
            rr = r + dr
            ss = s + ds

            if rr in data.index and str(ss) in data.columns:
                styles.loc[rr, str(ss)] = "background-color: lightgreen"

    return styles

df.style.apply(vybarvi_3x3, radek=radek, sloupec=sloupec, axis=None)
```
